```python
#importing relevant packages and loading data files. 

import pandas as pd
import numpy as np
Assesment = pd.read_csv("Assesment.csv")
Table_1 = pd.read_csv("Table_1.csv")
Table_2 = pd.read_csv("Table_2.csv")
```

```python
#filtering the data by highest days since joining.

Assesment_filtered = Assesment[Assesment["DAYS_SINCE_ACQUISITION"] == 168]
```

```python
Assesment_filtered
```

```python
#Setting the index for table 1
Table_1.set_index('ACQUISITION_CHANNEL')
```

```python
#Setting the index for table 2
Table_2.set_index('LOCATION')
```

```python
#Flattening the matrix into a multi index table.
CPA_Table = Table_1.set_index("ACQUISITION_CHANNEL").unstack().to_frame("CPA")

CPA_Table
```

```python
#Flattening the matrix into a multi index table.
Table_3 = Table_2.set_index("LOCATION").unstack().to_frame("Gross Margin per order")
```

```python
#Removing the multi index
Table_4 = Table_3.reset_index(level = 0, drop = False)
Table_4 = Table_4.rename(columns = {"level_0":"Date"})
Table_4
```

```python
#Extracting the week and year from the date.
Table_4["Date"] = pd.to_datetime(Table_4["Date"])

for date in Table_4["Date"]:
    Table_4["First_Work_Week"] = Table_4["Date"].dt.week

for date in Table_4["Date"]:
    Table_4["First_Work_Year"] = Table_4["Date"].dt.year
```

```python
#Creating a combined colunm for week-year to use for merging tables later. 
Table_4["Week-Year"] = Table_4["First_Work_Week"].astype(str) + Table_4["First_Work_Year"].astype(str)
Table_4
```

```python
#Merging the cPA table with the data table 
merge_1 = Assesment_filtered.merge(CPA_Table, left_on=["LOCATION","ACQUISITION_CHANNEL"], right_index=True, how="left")

merge_1

```

```python
#chnaging the data type for first work date to a date format
merge_1["FIRST_WORK_DATE"] = pd.to_datetime(merge_1["FIRST_WORK_DATE"]).dt.strftime('%d-%m-%y')
merge_1.dtypes
```

```python
#extracting the week and year from the merge_1 table
merge_1["FIRST_WORK_DATE"] = pd.to_datetime(merge_1["FIRST_WORK_DATE"])

for date in merge_1["FIRST_WORK_DATE"]:
    merge_1["First_Work_Week"] = merge_1["FIRST_WORK_DATE"].dt.week

for date in merge_1["FIRST_WORK_DATE"]:
    merge_1["First_Work_Year"] = merge_1["FIRST_WORK_DATE"].dt.year
```

```python
#combine the week-year colunm
merge_1["Week-Year"] = merge_1["First_Work_Week"].astype(str) + merge_1["First_Work_Year"].astype(str)
merge_1
#merge_2 = merge_1.merge(Table_2.set_index("LOCATION").unstack().to_frame("Gross Margin per order"), left_on=["LOCATION","FIRST_WORK_DATE"], right_index=True, how="left")



```

```python
#Second merge merging the merge_! table with the "Table_4" to get the gross margin per order.
merge_2 = pd.merge(merge_1,Table_4, on = ["Week-Year",'LOCATION', "First_Work_Week","First_Work_Year"],how = 'inner')

```

```python
merge_2
```

```python
#Calculating the gross margin per rider by multiplying the gross margin per order with the number of orders delivered.
merge_2["Gross Margin per rider"] = merge_2["Gross Margin per order"]*merge_2["ORDERS_DELIVERED_CUMULATIVE"]

merge_2
```
